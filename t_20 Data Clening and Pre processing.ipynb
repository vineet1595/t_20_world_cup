```python
# import Relevant libraries

import pandas as pd
import json
```

```python
# Importing the match summary file

with open('t20_json_files/t20_wc_match_results.json') as f:
        data = json.load(f)
data
```

```python
# checking the length of the array
len(data)
```

```python
data[0]['matchSummary']
```

```python
# transforming our array to Dataframe
df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
# Checking the size of our dataframe
df_match.shape
```

```python
# Changing the column header to atch_id to create a primary key
df_match.rename({'scorecard' : 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
#creating a dictionary to save the team1 Vs team2 & team2 VS team1 so that it can be further used as secondary key

match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']

    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']

match_ids_dict
```

** Batting Summary**

```python
# opening the batting summary json file

with open('t20_json_files/t20_wc_batting_summary.json') as f:
        data = json.load(f)

# Creating 1 list called all_records and the appending the records one by one using the extend method inside a for loop
        all_records = []

        for rec in data:
            all_records.extend(rec['battingSummary'])
        
all_records
```

```python
# transforming our array to Dataframe
df_batting = pd.DataFrame(all_records)
df_batting.head()
```

```python
# Changing the dismissal column into out/not out. If there is blank in the column that means the player was not out

df_batting['out/not out'] = df_batting.dismissal.apply(lambda x: 'out' if len(x) > 0 else 'not out')
df_batting
```

```python
# Dropping the dismisal column
df_batting.drop(columns = ['dismissal'], inplace = True)
df_batting.head(10)
```

```python
#Changing the setting to display all rows to check any special characters
pd.set_option('display.max_rows', None)
df_batting
```

```python
# Replacing the special characters with blanks â€
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€',''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0',''))
df_batting
```

```python
# creating a new column in batting dataframwe to add the match id for the respective match suing the .map function

df_batting['match_id'] = df_batting['match'].map(match_ids_dict)
df_batting.head(30)
```

```python
df_batting.to_csv('vineet project/batting_summary.csv', index = False)
```
